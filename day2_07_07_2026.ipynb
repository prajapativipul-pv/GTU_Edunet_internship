```python
#DICTIONARY
student = {"name": "Rahul", "age": 21}

# let()
print(student.get("name"))

# keys()
print(student.keys())

# values()
print(student.values())

# update()
student.update({"city": "Ahmedabad"})
print(student)

# pop()
student.pop("age")
print(student)
```

```python
#LIST
numbers = [100, 200, 300]

# 1. append()
numbers.append(40)

# 2. insert()
numbers.insert(1, 15)

# 3. remove()
numbers.remove(200)

# 4. pop()
numbers.pop()

# 5. sort()
numbers.sort()

print(numbers)
```

```python
#TUPLE
t = (10, 20, 30, 20)

print(len(t))        # Length
print(t.count(20))   # Count
print(t.index(30))   # Index
print(max(t))        # Maximum
print(min(t))        # Minimum
```

```python
#SET
s = {1, 2, 3}

# 1. add()
s.add(4)

# 2. remove()
s.remove(2)

# 3. discard()
s.discard(5)

# 4. pop()
s.pop()

# 5. clear()
s.clear()

print(s)
```

```python
#IF Statement
num = 10

if num > 0:
    print("Positive Number")
```

```python
#if-else Statement
num = 5

if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
#if-elif-else Statement
marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 60:
    print("Grade B")
else:
    print("Grade C")
```

```python
#Nested if-else
age = 20

if age >= 18:
    if age >= 21:
        print("Can Vote and Drive")
    else:
        print("Can Vote")
else:
    print("Not Eligible")
```

```python
#Break
for i in range(1, 11):
    if i == 5:
        break
    print(i)
```

```python
#Continue
for i in range(1, 11):
    if i == 5:
        continue
    print(i)
```

```python
#Pass
for i in range(1, 4):
    if i == 2:
        pass
    print(i)
```

```python
#for Loop
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```
