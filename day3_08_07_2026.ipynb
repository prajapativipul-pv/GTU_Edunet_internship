```python
#multidimensional array
import numpy as np

arr = np.array([[10, 20, 30],
                [40, 50, 60],
                [70, 80, 90]])

print("Original Array:")
print(arr)

# Access values using index numbers
print("\nAccess Elements:")
print("arr[0][0] =", arr[0][0])
print("arr[1][2] =", arr[1][2])
print("arr[2][1] =", arr[2][1])

# Mathematical operations
sum_val = arr[0][0] + arr[1][1]
product = arr[0][2] * arr[2][0]
difference = arr[2][2] - arr[0][1]
division = arr[2][1] / arr[1][0]
print("\nMathematical Operations:")
print("Addition =", sum_val)
print("Multiplication =", product)
print("Subtraction =", difference)
print("Division =", division)


print("\n1. Shape:", arr.shape)
print("2. Size:", arr.size)
print("3. Dimensions:", arr.ndim)
print("4. Data Type:", arr.dtype)
print("5. Reshape:\n", arr.reshape(1,9))
print("6. Flatten:", arr.flatten())
print("7. Transpose:\n", arr.T)
print("8. Maximum:", np.max(arr))
print("9. Minimum:", np.min(arr))
print("10. Sum:", np.sum(arr))
print("11. Mean:", np.mean(arr))
print("12. Standard Deviation:", np.std(arr))
print("13. Sorted Array:\n", np.sort(arr))
print("14. Square Root:\n", np.sqrt(arr))
print("15. Unique Values:", np.unique(arr))
```
