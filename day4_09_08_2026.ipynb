```python
# Sustainability Dataset Analysis
# Dataset: Our World in Data Energy Dataset
# Source: https://owid-public.owid.io/data/energy/owid-energy-data.csv

import pandas as pd
import numpy as np


# Placeholder for a working dataset
try:
    # Attempt to load the original dataset URL again
    url = "https://owid-public.owid.io/data/energy/owid-energy-data.csv"
    df = pd.read_csv(url)
except Exception as e:
    print(f"Could not load from original URL: {e}. Please ensure the file is downloaded and loaded locally.")

    print("Loading a sample dataset for demonstration purposes. Please replace with 'owid-energy-data.csv' once uploaded.")
    data = {
        'country': ['USA', 'China', 'India', 'USA', 'China', 'India'],
        'year': [2019, 2019, 2019, 2020, 2020, 2020],
        'primary_energy_consumption': [100, 200, 50, 105, 210, 55],
        'gdp': [21000, 14000, 2900, 21500, 14500, 3000]
    }
    df = pd.DataFrame(data)

# Basic Exploration
# View first rows
print(df.head())

# Dataset shape
print("Rows, Columns:", df.shape)

# Data types
print(df.dtypes)

# Statistical summary
print(df.describe(include='all'))

# Missing Value Handling

# Numeric columns -> replace with median
num_cols = df.select_dtypes(include=np.number).columns
# Check if num_cols is not empty before applying median fillna
if not num_cols.empty:
    df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Categorical columns -> replace with mode
cat_cols = df.select_dtypes(exclude=np.number).columns
for col in cat_cols:
    # Check if mode() returns values before attempting to index
    mode_val = df[col].mode()
    if not mode_val.empty:
        df[col] = df[col].fillna(mode_val[0])

print("Remaining null values:")
print(df.isnull().sum())



# Simple Analysis

print("\nCorrelation among numeric features:")
# Check if num_cols is not empty before calculating correlation
if not num_cols.empty and len(num_cols) > 1:
    print(df[num_cols].corr())
else:
    print("Not enough numeric features for correlation calculation.")

print("\nTop 10 countries by latest primary energy consumption:")
# Ensure 'country' and 'primary_energy_consumption' exist in df columns
if "country" in df.columns and "primary_energy_consumption" in df.columns:
    latest=df.groupby("country").tail(1)
    print(latest.sort_values("primary_energy_consumption",ascending=False)[["country","primary_energy_consumption"]].head(10))
else:
    print("Required columns 'country' or 'primary_energy_energy_consumption' not found for this analysis.")


# SUMMARY

print("""
Summary:
1. Downloaded a sustainability (energy) dataset.
2. Imported dataset using pandas.
3. Explored structure, statistics and data types.
4. Documented feature relationships in comments.
5. Handled missing values:
   - Numeric -> Median
   - Categorical -> Mode
6. Calculated correlation matrix.
7. Displayed top countries by energy consumption (if available).
""")
```
